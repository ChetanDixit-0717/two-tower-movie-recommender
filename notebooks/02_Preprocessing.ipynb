# Import Required Libraries

Import the essential Python libraries required for data manipulation and analysis.

```python
import pandas as pd
import numpy as np

pd.set_option("display.max_columns", None)
pd.set_option("display.max_colwidth", None)
```

# Load the Dataset

Load the MovieLens datasets containing movie information, user ratings, and user metadata.

```python
movies = pd.read_csv("../data/movies.csv")
ratings = pd.read_csv("../data/ratings.csv")
users = pd.read_csv("../data/users.csv")
```

# Select Relevant Features

Keep only the columns required for the baseline Two-Tower recommendation model. This reduces memory usage and simplifies preprocessing.

```python
ratings = ratings[
    ["userId", "movieId", "rating"]
]

movies = movies[
    ["movieId", "title"]
]
```

# Merge Ratings with Movie Information

Combine user ratings with movie titles using the common `movieId` column to create a unified dataset for model training.

```python
data = ratings.merge(
    movies,
    on="movieId",
    how="inner"
)
```
